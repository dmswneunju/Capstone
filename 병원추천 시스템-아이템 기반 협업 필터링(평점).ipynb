### 2. 사용자별 병원 평점 파일 생성

```python
import numpy as np
import pandas as pd

# 600명의 사용자 id 생성
user_ids = np.arange(1, 601)

# 1364개의 병원 id 생성
hospital_ids = np.arange(1, 1365)

# 각 사용자별로 랜덤하게 10개의 병원 선택
selected_hospitals = np.random.choice(hospital_ids, size=(600, 10), replace=True) #복원추출

# 선택된 10개의 병원에 대해 1~5사이의 랜덤한 평점 부여
ratings = np.random.randint(1, 6, size=(600, 10))

# 데이터프레임 생성
df = pd.DataFrame(data=np.column_stack((np.repeat(user_ids, 10), selected_hospitals.ravel(), ratings.ravel())),
                  columns=['user_id', 'id', 'rating'])
df
```

```python
# csv 파일로 저장
df.to_csv('ratings.csv', index=False)
```

## 아이템 기반 최근접 이웃 협업 필터링
* 사용자 기반 : 각 사용자와 비슷한 고객들의 행동 참고
* 아이템 기반 : 해당 아이템을 선택한 다른 고객들의 행동 참고
* 일반적으로는 사용자 기반보다 아이템 기반 협업 필터링이 정확도가 높다.  
    * 비슷한 병원을 선호했다고 사용자의 병원선택 기준이 비슷하다고 판단하기 어렵거나 매우 유명한 병원의 경우 선호하는 병원과 관계없이 선택하는 경우가 많고 평점을 매기지 않는 경우가 많기 때문이다.

### 병원과 평점 데이터를 병원 id기준으로 병합

```python
rating_hospital = pd.merge(meta1, df, on='id')
rating_hospital.head(2)
```

### rating데이터를 각 사용자의 평점을 보기 쉽게 pivot_table로 변경

```python
ratings_matrix = rating_hospital.pivot_table('rating', index='user_id', columns='name')
ratings_matrix.fillna(0, inplace=True)
ratings_matrix.head(2)
```

### 유사도 측정
유사도 측정 위해 행렬 transpose(행, 열 전환)

```python
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T.head(2)
```

### 코사인 유사도 측정

!pip install -U scikit-learn

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ratings_matrix_T, ratings_matrix_T)
item_sim_df = pd.DataFrame(data=item_sim, index=ratings_matrix.columns, columns=ratings_matrix.columns)

print(item_sim_df.shape)
item_sim_df.head(2)
```

### 추천 병원 dataframe반환 함수
해당병원과 평점이 비슷한 병원 추천

```python
def find_sim_hospital_item(df, hospital_name, top_n=10):
    name_hospital_sim = df[[hospital_name]].drop(hospital_name, axis=0) #입력한 병원을 drop

    return name_hospital_sim.sort_values(hospital_name, ascending=False)[:top_n]
```

```python
find_sim_hospital_item(item_sim_df, '새빛요양병원')
```
