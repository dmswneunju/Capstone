### 1. 병원데이터에서 필요한부분만 가져오기  

```python
import pandas as pd
meta = pd.read_csv('전국요양병원api.csv')
```

```python
meta #1364 rows × 13 columns
```

```python
meta = meta.dropna(subset=['category'])#카테고리 없는 행만 삭제
meta.head(5)
```

```python
meta1 = meta.drop(['photo','open_time','credit','createdat','updatedat'],axis=1) # 필요없는 column제거
```

```python
meta1
```

## 병원 id 재정렬

```python
meta1['id']
```

```python
sum(meta1.duplicated(subset=['id'], keep=False))#id에는 중복값 없음.
```

```python
meta1['id'] = [i+1 for i in range(len(meta['id']))]
```

```python
meta1 #병원별 id -> rename
```

```python
meta1.to_csv('hospital.csv', index=False)
```
