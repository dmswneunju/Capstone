```python
df5 = pd.read_csv('hospital_pharmacy(23.03)/5.의료기관별상세정보서비스_03_진료과목정보.csv')
df5['종별코드']=28
```

```python
df5 #370866 rows × 7 columns
```

```python
df5.info()
```

```python
df5['진료과목코드명'].value_counts().head(10)
```

```python
df = pd.merge(left=df1, # 왼쪽 DataFrame 지정
              right=df5, # 오른쪽 DataFrame 지정
              how='left', # 조인 방법 지정
              left_on='암호화요양기호', # 왼쪽 DataFrame에서 기준으로 삼을 열 지정
              right_on='암호화요양기호')
```

```python
df #373914 rows × 36 columns
```

```python
df['진료과목코드'].isna().sum()
```

```python
df.info()
```

```python
df_code = df.dropna(subset=['진료과목코드']) #진료과목코드 없는 병원 삭제
```

```python
df_code
```

```python
df_code.info()
```

```python
df_code=df_code.drop(['요양기관명_y', '종별코드_y','읍면동'], axis=1) #중복 열, 불필요햔 열 제거
```

```python
df_code=df_code.rename(columns={'요양기관명_x':'요양기관명'})
```

```python
df_code=df_code.rename(columns={'종별코드_x':'종별코드'})
```

```python
df_code.info()
```

```python
#좌표없는 병원 제거
df_code = df_code.dropna(subset=['좌표(X)']) #370535개의 열로 줄임.
```
